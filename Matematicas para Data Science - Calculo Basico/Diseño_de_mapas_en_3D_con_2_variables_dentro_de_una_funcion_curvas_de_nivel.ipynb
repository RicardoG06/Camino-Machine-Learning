```python
from matplotlib import cm #Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Grafica de 3 dimensiones

def Z(x,y):
  return np.sin(x) + 2*np.cos(y)

res = 100

x = np.linspace(-4,4,num = res)

y = np.linspace(-4,4,num = res)

x , y = np.meshgrid(x,y) 

#Sirve para poder meter todos los datos de X e Y dentro de arrays (vs de X e Y)
#Ejemplo:
#coord_x = [0,1,2]
#coord_y = [0,1,2]
#x , y = np.meshgrid(coord_x,coord_y) 
#print(x)
#[[0,1,2]
# [0,1,2]
# [0,1,2]]
#print(y)
#[[0,0,0]
# [1,1,1]
# [2,2,2]]

z = Z(x,y)

fig , ax = plt.subplots(subplot_kw={"projection":"3d"}) 
#Realizara el tipo proyeccion en 3d

superficie = ax.plot_surface(x,y,z, cmap = cm.cool)
fig.colorbar(superficie)
```

```python
#Curvas de nivel
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(z),np.max(z),num = res)

cp = ax2.contour(x,y,z,levels = level_map,cmap = cm.cool)
fig2.colorbar(cp)
```
