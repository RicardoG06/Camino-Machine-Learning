```python
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas
```

```python
# F(X) = mx + b

def f(m,x,b):
  return m*x + b

res = 100 #Resolucion de la grafica

m = 1

b = 5

x = np.linspace(-10 , 10 , num = res) #Genera una serie de puntos en un rango

y = f(m,x,b)

fig , ax = plt.subplots()

ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color ='r')
ax.axvline(x=0, color ='r')
```

```python
# F(X) = 2x^8 - x^4 + 3x^3 + 4

def f(x):
  return 2*x**8 - x**4 + 3*x**3 + 4

res = 100 #Resolucion de la grafica

x = np.linspace(-10 , 10 , num = res) #Genera una serie de puntos en un rango

y = f(x)

fig , ax = plt.subplots()
plt.xlim(-10, 10) #Definir el tamaño del 'x' y 'y' en la graf
plt.ylim(0, 10)
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color ='r')
ax.axvline(x=0, color ='r')
```

```python
# F(x) = {0 , para x < 0}
#        {1 , para x >= 0}

def f(x):
  y = np.zeros(len(x))
  for idx , x in enumerate(x):
    if x >= 0:
      y[idx] = 1
  return  y


res = 1000 #Resolucion de la grafica

x = np.linspace(-10 , 10 , num = res) #Genera una serie de puntos en un rango
y = f(x)

fig , ax = plt.subplots()
ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color ='r')
ax.axvline(x=0, color ='r')
```
